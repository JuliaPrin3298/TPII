# 🧠 O que são Testes de Software?

Testes são procedimentos utilizados para verificar se o sistema
funciona conforme esperado.

Eles ajudam a identificar:

- erros
- falhas de lógica
- comportamentos inesperados



# 📚 Tipos de Testes

Existem vários tipos de testes.

### Testes Unitários

Testam pequenas partes do sistema, como funções.

### Testes de Integração

Testam interação entre diferentes partes do sistema.

### Testes de Sistema

Testam o sistema completo.



Nesta aula vamos focar em:

## Testes Unitários



# ⚙ Biblioteca de Testes

Python possui a biblioteca padrão:

```
unittest
```


````python
import unittest

def soma(a, b):
    return a + b

class TesteSoma(unittest.TestCase):

    def test_soma(self):
        resultado = soma(2,3)
        self.assertEqual(resultado,5)

if __name__ == "__main__":
    unittest.main()
````


Esse teste verifica se a função **soma()** funciona corretamente.



# 🧪 Testando o Modelo Produto

Podemos testar também nosso modelo.


````python
def criar_produto(nome, preco):
    if nome == "":
        raise ValueError("Nome obrigatório")
    return {"nome": nome, "preco": preco}
````

````python
class TestProduto(unittest.TestCase):

    def test_criar_produto(self):
        produto = criar_produto("Notebook",3000)
        self.assertEqual(produto["nome"],"Notebook")
````


# 🧪 Executando os Testes

Para executar testes:

```
python -m unittest
```
