# 🔁 Revisão

Relembrando a aula anterior:

- Criamos uma aplicação Flask
- Subimos o servidor local
- Criamos a primeira rota `/`

Agora vamos evoluir esse sistema.


```python

from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "Sistema de Produtos"

if __name__ == "__main__":
    app.run(debug=True)
```


# 🧱 BLOCO 1 — Estruturando o Sistema

Vamos pensar no sistema como um produto real.

Quais páginas precisamos?

- Produtos
- Clientes
- Categorias
- Pedidos

Agora vamos implementar isso.


```python

@app.route("/produtos")
def produtos():
    return "Lista de produtos"

@app.route("/clientes")
def clientes():
    return "Lista de clientes"

@app.route("/categorias")
def categorias():
    return "Lista de categorias"

@app.route("/pedidos")
def pedidos():
    return "Lista de pedidos"
```

# 🧠 BLOCO 2 — Rotas Dinâmicas

Agora vamos tornar o sistema dinâmico.

Problema atual:

👉 Não conseguimos acessar um produto específico.

Solução:

👉 Usar parâmetros na URL.


```python

@app.route("/produto/<nome>")
def produto(nome):
    return f"Produto selecionado: {nome}"
```

# 🔢 BLOCO 3 — Tipagem de Parâmetros

Agora vamos evitar erros.

Exemplo: ID deve ser número.


```python

@app.route("/pedido/<int:id>")
def pedido(id):
    return f"Pedido número: {id}"
```


## Tipos disponíveis

| Tipo   | Descrição          |
|--------|--------------------|
| string | Texto              |
| int    | Número inteiro     |
| float  | Número decimal     |
| path   | Caminho completo   |



# ⚠️ ERRO COMUM (IMPORTANTÍSSIMO)

Errado:
