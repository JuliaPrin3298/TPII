# 🧠 Por que documentar um sistema?

A documentação ajuda:

- novos desenvolvedores a entender o projeto
- usuários a utilizar o sistema
- equipes a manter o software
- empresas a registrar funcionamento do sistema

Sem documentação, um projeto pode se tornar **difícil de manter**.



# 📚 Tipos de Documentação

Existem diferentes tipos de documentação:

### Documentação Técnica
Explica arquitetura e funcionamento do sistema.

### Documentação de Usuário
Explica como utilizar o sistema.

### Documentação de API
Explica como integrar com o sistema.



# 📄 README.md

O arquivo mais comum em projetos é o:

```
README.md
```
