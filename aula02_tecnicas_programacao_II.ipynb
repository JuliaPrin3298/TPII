# 🌐 O que é uma aplicação web?

Uma aplicação web é um sistema que funciona através de um navegador.

Exemplos:

- sistemas acadêmicos
- e-commerce
- redes sociais
- dashboards corporativos

Arquitetura básica:

```
Navegador → Servidor Web → Aplicação → Banco de Dados
```



# ⚙ Backend vs Frontend

## Frontend

Responsável pela interface do usuário.

Tecnologias comuns:

- HTML
- CSS
- JavaScript

## Backend

Responsável pela lógica do sistema.

Tecnologias comuns:

- Python
- Java
- Node.js
- PHP



# 🧠 O que é um Framework?

Framework é uma estrutura que fornece ferramentas e padrões para facilitar o desenvolvimento de software.

Vantagens:

- organização do código
- reutilização de componentes
- aumento da produtividade
- padronização da arquitetura



# 🐍 O que é Flask?

Flask é um **microframework web para Python**.

Características:

- simples
- leve
- flexível
- ideal para APIs e aplicações web
- muito utilizado em projetos educacionais e startups


# 🧱 Estrutura mínima de uma aplicação Flask


````python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "Olá Flask!"

if __name__ == "__main__":
    app.run(debug=True)
````


# 🔎 Explicando o código

### Flask()

Cria a aplicação web.

### @app.route()

Define uma rota da aplicação.

### app.run()

Inicializa o servidor web.



# 🚀 Executando a aplicação

Execute o arquivo:


Abra no navegador:

# 🧪 Exemplo com múltiplas rotas


````python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "Página inicial"

@app.route("/sobre")
def sobre():
    return "Página sobre o sistema"

@app.route("/produtos")
def produtos():
    return "Lista de produtos"

if __name__ == "__main__":
    app.run(debug=True)
````
