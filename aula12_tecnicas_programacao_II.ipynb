# 🧠 O que são Design Patterns?

**Design Patterns** são soluções reutilizáveis para problemas comuns no desenvolvimento de software.

Eles não são códigos prontos, mas **formas estruturadas de organizar o código**.



# 📚 Origem dos Design Patterns

O conceito ficou famoso com o livro:

**Design Patterns: Elements of Reusable Object-Oriented Software**

Autores:

- Erich Gamma
- Richard Helm
- Ralph Johnson
- John Vlissides

Eles ficaram conhecidos como:

### Gang of Four (GoF)



# 🧩 Tipos de Design Patterns

Os padrões são divididos em três grupos principais:

### 1️⃣ Padrões de Criação

Controlam a criação de objetos.

Exemplos:

- Singleton
- Factory
- Builder



### 2️⃣ Padrões Estruturais

Organizam a estrutura entre classes.

Exemplos:

- Adapter
- Decorator
- Facade



### 3️⃣ Padrões Comportamentais

Controlam comunicação entre objetos.

Exemplos:

- Observer
- Strategy
- Command



# 🧱 Exemplo de Problema

Imagine um sistema onde várias partes do código acessam o banco diretamente.

Exemplo:


````python
produtos = Produto.query.all()

produto = Produto.query.get(1)

db.session.add(produto)
db.session.commit()
````


Quando o sistema cresce, isso pode gerar:

- código duplicado
- dificuldade de manutenção
- dificuldade de testes



# 🚀 Solução com Design Pattern

Uma solução comum é utilizar o padrão:

## Repository Pattern

Ele cria uma camada responsável por **acessar o banco de dados**.


````python
class ProdutoRepository:

    def listar(self):
        return Produto.query.all()

    def buscar_por_id(self, id):
        return Produto.query.get(id)

    def salvar(self, produto):
        db.session.add(produto)
        db.session.commit()
````


Agora o controller pode usar o repositório:


````python
repo = ProdutoRepository()

produtos = repo.listar()
````


Isso melhora:

- organização
- reutilização
- manutenção do sistema
