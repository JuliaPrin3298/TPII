# ✔ O que é Validação?

Validação é o processo de **verificar se os dados enviados são válidos antes de salvar no banco**.

Exemplos de validação:

- campo obrigatório
- tamanho mínimo
- valor numérico
- valor positivo



# 🧠 Validação no Controller


````python
@app.route("/novo", methods=["GET","POST"])
def novo_produto():

    if request.method == "POST":

        nome = request.form["nome"]
        preco = request.form["preco"]

        if nome == "":
            return "Erro: nome do produto é obrigatório"

        if preco == "":
            return "Erro: preço é obrigatório"

        produto = Produto(nome=nome, preco=preco)

        db.session.add(produto)
        db.session.commit()

        return redirect("/produtos")

    return render_template("novo_produto.html")
````


Observe que verificamos se os campos estão vazios antes de salvar.



# 💬 Mensagens para o Usuário

# 🚀 Utilizando Flash Messages

Flask possui um recurso chamado:

````python
from flask import flash

flash("Produto cadastrado com sucesso!")
````


# 📄 Exibindo Mensagens no Template


````python
{% with messages = get_flashed_messages() %}
{% if messages %}

{% for message in messages %}

<div class="alert alert-success">
{{ message }}
</div>

{% endfor %}

{% endif %}
{% endwith %}
````


Essas mensagens aparecerão na interface do sistema.
