# ⚠ Problema na Criação de Objetos

Em sistemas grandes, objetos podem ser criados em vários lugares.

Exemplo:


````python
produto = Produto(nome="Notebook", preco=3500)

produto = Produto(nome="Mouse", preco=80)

produto = Produto(nome="Teclado", preco=150)
````


Se a forma de criação mudar (por exemplo adicionar validações),
será necessário alterar **vários pontos do sistema**.



# 🚀 Solução: Factory Pattern

O **Factory Pattern** cria uma classe responsável por **criar objetos**.

Assim centralizamos a lógica de criação.



# 📂 Estrutura do Projeto

```
projeto/

models/
    produto.py

repositories/
    produto_repository.py

factories/
    produto_factory.py
```



# 🧱 Criando a Factory


````python
# factories/produto_factory.py

class ProdutoFactory:

    @staticmethod
    def criar(nome, preco):

        if nome == "":
            raise ValueError("Nome do produto é obrigatório")

        if float(preco) <= 0:
            raise ValueError("Preço deve ser positivo")

        return Produto(nome=nome, preco=preco)
````


Agora toda criação de produto passa pela **factory**.



# 🧠 Utilizando a Factory no Controller


````python
from factories.produto_factory import ProdutoFactory

@app.route("/novo", methods=["GET","POST"])
def novo_produto():

    if request.method == "POST":

        nome = request.form["nome"]
        preco = request.form["preco"]

        produto = ProdutoFactory.criar(nome, preco)

        repo.salvar(produto)

        return redirect("/produtos")

    return render_template("novo_produto.html")
````


Observe que agora o controller **não cria o objeto diretamente**.
