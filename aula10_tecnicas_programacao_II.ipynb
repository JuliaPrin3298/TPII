# 🧠 O que são Exceções?

Exceções são **erros que ocorrem durante a execução do programa**.

Exemplo simples em Python:


````python
numero = int(input("Digite um número: "))
print(numero)
````


Se o usuário digitar um texto, ocorrerá um erro.

Podemos tratar esse erro utilizando:

```
try / except
```


````python
try:
    numero = int(input("Digite um número: "))
    print(numero)
except:
    print("Erro: valor inválido")
````


# 🚀 Tratamento de Erros no Flask

Flask permite criar **páginas de erro personalizadas**.

Exemplo:

Erro **404** ocorre quando uma página não existe.


````python
@app.errorhandler(404)
def pagina_nao_encontrada(e):
    return render_template("404.html"), 404
````


# 📄 Criando Página 404

Crie o arquivo:

```
templates/404.html
```


````python
{% extends "base.html" %}

{% block conteudo %}

<h1>Página não encontrada</h1>

<p>A página que você tentou acessar não existe.</p>

<a href="/">Voltar para página inicial</a>

{% endblock %}
````


Agora sempre que uma página inexistente for acessada,
o sistema mostrará **uma página personalizada**.



# 🧠 Tratando Erros no Banco de Dados

Podemos usar **try / except** ao salvar dados.


````python
try:

    produto = Produto(nome=nome, preco=preco)

    db.session.add(produto)
    db.session.commit()

    flash("Produto salvo com sucesso")

except Exception as e:

    db.session.rollback()

    flash("Erro ao salvar produto")
````


Isso evita que o sistema quebre caso ocorra algum erro.
