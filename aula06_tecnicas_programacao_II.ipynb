# 🧠 O que é um Banco de Dados?

Um banco de dados é um sistema utilizado para **armazenar e organizar informações**.

Exemplos de bancos de dados:

- MySQL
- PostgreSQL
- SQLite
- SQL Server
- MongoDB



# 📊 Banco de Dados Relacional

Bancos de dados relacionais armazenam dados em **tabelas**.

Exemplo:

| id | nome | preco |
|----|------|-------|
| 1 | Mouse | 50 |
| 2 | Teclado | 120 |
| 3 | Monitor | 900 |



# 🧠 O que é ORM?

ORM significa:

**Object Relational Mapping**

Ele permite trabalhar com banco de dados utilizando **objetos Python**.

Ou seja:

Em vez de escrever SQL diretamente, podemos utilizar **classes Python**.



# 🐍 SQLAlchemy

SQLAlchemy é um dos ORMs mais utilizados em Python.

Ele permite:

- mapear classes para tabelas
- inserir dados
- consultar dados
- atualizar registros
- remover registros



# 📦 Instalando SQLAlchemy


````python
pip install flask_sqlalchemy
````


# ⚙ Configurando SQLAlchemy no Flask


````python
from flask import Flask
from flask_sqlalchemy import SQLAlchemy

app = Flask(__name__)

app.config['SQLALCHEMY_DATABASE_URI'] = 'sqlite:///produtos.db'
app.config['SQLALCHEMY_TRACK_MODIFICATIONS'] = False

db = SQLAlchemy(app)
````


# 🧱 Criando o Modelo Produto

Agora vamos criar uma classe que representa a tabela **Produto**.


````python
class Produto(db.Model):

    id = db.Column(db.Integer, primary_key=True)
    nome = db.Column(db.String(100))
    preco = db.Column(db.Float)
````


Essa classe gera automaticamente uma tabela no banco:

```
Produto
```



# 🏗 Criando o Banco de Dados


````python
with app.app_context():
    db.create_all()
````


Depois de executar esse código, o arquivo:

```
produtos.db
```

será criado no projeto.



# ➕ Inserindo dados no banco


````python
produto = Produto(nome="Mouse", preco=50)

db.session.add(produto)
db.session.commit()
````


# 🔎 Consultando dados


````python
produtos = Produto.query.all()

for p in produtos:
    print(p.nome, p.preco)
````


# 🔍 Buscando um registro específico


````python
produto = Produto.query.get(1)

print(produto.nome)
````
